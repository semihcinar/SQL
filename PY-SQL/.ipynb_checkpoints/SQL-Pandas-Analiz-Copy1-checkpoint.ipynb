```python
olumlu=[]
olumluSonuc=[]

olumsuz=[]
olumsuzSonuc=[]

yorumsuz = []


for yazi in metin:
    blob1 = TextBlob(yazi)
    blob_eng = blob1.translate(to='en')
    if(blob_eng.polarity > 0) :
        olumlu.append(yazi)
        olumluSonuc.append(blob_eng.sentiment) 
    elif(blob_eng.polarity < 0) :
        olumsuz.append(yazi)
        olumsuzSonuc.append(blob_eng.sentiment)
    else:
        yorumsuz.append(yazi)
        
        
    
    
```
